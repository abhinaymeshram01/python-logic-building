```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def insert_at_beginning(head, data):
    new_node = Node(data)
    new_node.next = head
    return new_node


# Example
head = Node(20)
head.next = Node(30)

head = insert_at_beginning(head, 10)

current = head
while current:
    print(current.data, end=" -> ")
    current = current.next
```
