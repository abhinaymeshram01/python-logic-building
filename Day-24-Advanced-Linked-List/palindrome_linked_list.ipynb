```python
def is_palindrome(head):
    slow = head
    fast = head

    # Find middle
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

    # Reverse second half
    prev = None
    while slow:
        next_node = slow.next
        slow.next = prev
        prev = slow
        slow = next_node

    # Compare halves
    left = head
    right = prev

    while right:
        if left.data != right.data:
            return False
        left = left.next
        right = right.next

    return True
```
