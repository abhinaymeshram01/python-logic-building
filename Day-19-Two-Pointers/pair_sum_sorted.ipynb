```python
arr = list(map(int, input("Enter sorted numbers: ").split()))
target = int(input("Enter target: "))

left, right = 0, len(arr) - 1

while left < right:
    s = arr[left] + arr[right]
    if s == target:
        print(arr[left], arr[right])
        break
    elif s < target:
        left += 1
    else:
        right -= 1
else:
    print("No pair found")
```
