```python
arr = list(map(int, input("Enter sorted numbers: ").split()))

if not arr:
    print([])
else:
    slow = 0

    for fast in range(1, len(arr)):
        if arr[fast] != arr[slow]:
            slow += 1
            arr[slow] = arr[fast]

    print("Array after removing duplicates:", arr[:slow+1])
```
