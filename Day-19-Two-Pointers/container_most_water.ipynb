```python
arr = list(map(int, input("Enter heights: ").split()))
left, right = 0, len(arr) - 1
max_area = 0

while left < right:
    width = right - left
    height = min(arr[left], arr[right])
    max_area = max(max_area, width * height)

    if arr[left] < arr[right]:
        left += 1
    else:
        right -= 1

print("Maximum water:", max_area)
```
