```python
arr = list(map(int, input("Enter elements: ").split()))
freq = {}

for num in arr:
    freq[num] = freq.get(num, 0) + 1

max_elem = arr[0]
max_count = freq[max_elem]

for key in freq:
    if freq[key] > max_count:
        max_elem = key
        max_count = freq[key]

print("Element:", max_elem)
print("Frequency:", max_count)
```
