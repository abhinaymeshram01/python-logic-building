```python
arr = list(map(int, input("Enter sorted numbers: ").split()))
target = int(input("Enter target: "))

left, right = 0, len(arr) - 1
result = -1

while left <= right:
    mid = (left + right) // 2

    if arr[mid] == target:
        result = mid
        right = mid - 1
    elif arr[mid] < target:
        left = mid + 1
    else:
        right = mid - 1

print("First occurrence:", result)
```
