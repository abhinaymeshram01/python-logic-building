```python
class Calculator:
    def add(self, a, b):
        return a + b

    def sub(self, a, b):
        return a - b

    def mul(self, a, b):
        return a * b

    def div(self, a, b):
        if b == 0:
            return "Division by zero not allowed"
        return a / b


calc = Calculator()
print(calc.add(10, 5))
print(calc.div(10, 0))
```
