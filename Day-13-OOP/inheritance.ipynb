```python
class Person:
    def __init__(self, name):
        self.name = name

    def show(self):
        print("Name:", self.name)


class Employee(Person):
    def __init__(self, name, emp_id):
        super().__init__(name)
        self.emp_id = emp_id

    def show_details(self):
        self.show()
        print("Employee ID:", self.emp_id)


e = Employee("Abhi", 101)
e.show_details()
```
