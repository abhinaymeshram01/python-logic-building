```python
def is_palindrome(s, left, right):
    if left >= right:
        return True
    if s[left] != s[right]:
        return False
    return is_palindrome(s, left + 1, right - 1)
```

```python
s = input("Enter string: ")
print("Palindrome" if is_palindrome(s, 0, len(s) - 1) else "Not Palindrome")
```
