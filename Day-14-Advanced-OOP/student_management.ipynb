```python
class Student:
    def __init__(self, roll, name, marks):
        self.roll = roll
        self.name = name
        self.marks = marks

    def __str__(self):
        return f"{self.roll} | {self.name} | {self.marks}"


students = []

def add_student():
    roll = int(input("Roll: "))
    name = input("Name: ")
    marks = int(input("Marks: "))
    students.append(Student(roll, name, marks))


def display_students():
    for s in students:
        print(s)


def search_student():
    roll = int(input("Enter roll to search: "))
    for s in students:
        if s.roll == roll:
            print(s)
            return
    print("Student not found")


while True:
    print("\n1.Add 2.Display 3.Search 4.Exit")
    ch = int(input("Enter choice: "))

    if ch == 1:
        add_student()
    elif ch == 2:
        display_students()
    elif ch == 3:
        search_student()
    elif ch == 4:
        break
    else:
        print("Invalid choice")
```
