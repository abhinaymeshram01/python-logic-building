```python
class Employee:
    company = "ABC Corp"   # class variable

    def __init__(self, name, salary):
        self.name = name        # instance variable
        self.salary = salary

    def display(self):
        print(self.name, self.salary, Employee.company)


e1 = Employee("Abhi", 50000)
e2 = Employee("Ravi", 60000)

e1.display()
e2.display()
```
