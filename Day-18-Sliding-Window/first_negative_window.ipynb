```python
from collections import deque

arr = list(map(int, input("Enter numbers: ").split()))
k = int(input("Enter window size: "))

q = deque()

for i in range(len(arr)):
    if arr[i] < 0:
        q.append(i)

    if i >= k - 1:
        if q and q[0] < i - k + 1:
            q.popleft()

        print(arr[q[0]] if q else 0, end=" ")
```
